```python
# Split the DataFrame
df1 = titanic_df.iloc[:500]  # First 500 rows
'''
1-So what is ILOC and is it a function first or is it a method?

"The iloc is a method in pandas, and it's used for indexing and selecting data by integer location.
Essentially, it allows you to access rows and columns by their numerical index positions rather than labels.
In the line df1 = titanic_df.iloc[:500], what we're doing is selecting the first 500 rows of the DataFrame.
The colon : before the 500 indicates that we want all rows from the start up to, but not including, the 500th row.
So, iloc is definitely a method, and it's super handy for positional indexing! If you have any other questions about it, let me know!"

2-Does it actually include the 500th row, or does it exclude it?

"The slicing in Python is exclusive of the end index.
That means that when you use iloc[:500], it will include rows from index 0 up to index 499.
The row at index 500 is not included.
So, effectively, you'll get the first 500 rows, with the last one being the 499th index."

3- Axis = 0 means to add row by row
https://tmlb-blog-ltd.medium.com/to-avoid-confusion-about-the-direction-of-an-axis-in-python-pandas-dcf466615731

4- "The reset_index method is used to reassign a new index to the dataframe,
and the drop=True argument means that the old index will not be added as a new column in the dataframe.
If you don't use drop=True, the old index will become a separate column,
which you usually don't need if you're just reassembling the dataframe.
So, drop=True keeps it nice and clean!"

'''

df2 = titanic_df.iloc[500:]  # Remaining rows

# Concatenate them back together
reconstructed_df = pd.concat([df1, df2], axis=0).reset_index(drop=True)

# Check if it's the same as the original
print(reconstructed_df.equals(my_df))

```
